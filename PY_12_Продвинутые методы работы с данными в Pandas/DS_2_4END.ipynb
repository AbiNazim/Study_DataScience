```python
import pandas as pd
import re 
```

```python
ratingsEND = pd.read_csv('data_END/ratings_movies.csv')
display(ratingsEND)
```

```python
#библиотека для регулярных выражений
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

```python
rat_END_copy = ratingsEND.copy()
```

```python
rat_END_copy['year_release'] = rat_END_copy['title'].apply(get_year_release)
display(rat_END_copy)
```

```python
rat_END_copy['year_release'].isnull().value_counts()  # Создайте в таблице новый признак year_release, который соответствует году выпуска фильма. У скольких фильмов не указан год их выпуска?
```

```python
mask = rat_END_copy['year_release'] == 1999
rat_END_copy[mask].groupby('title')['rating'].mean().sort_values()
# Какой фильм, выпущенный в 1999 году, получил наименьшую среднюю оценку зрителей? В качестве ответа запишите название этого фильма без указания года его выпуска.
```

```python
mask1 = rat_END_copy['year_release'] == 2010
rat_END_copy[mask1].groupby('genres')['rating'].mean().sort_values()
# Какое сочетание жанров фильмов (genres), выпущенных в 2010 году, получило наименьшую среднюю оценку (rating)?
```

```python
rat_END_copy.groupby('userId')['genres'].nunique().sort_values(ascending=False)
# Какой пользователь (userId) посмотрел наибольшее количество различных (уникальных) комбинаций жанров (genres) фильмов? В качестве ответа запишите идентификатор этого пользователя.
```

```python
rat_END_copy.groupby('userId')['rating'].agg(
    'min', 'mean'
).sort_values(ascending=False)
# Найдите пользователя, который выставил наименьшее количество оценок, но его средняя оценка фильмам наибольшая. В качестве ответа укажите идентификатор этого пользователя.
```

```python
mask2010 = rat_END_copy['year_release'] == 2018
maskGR = rat_END_copy[mask2010].groupby('genres')['rating'].agg(
    ['mean', 'count']
)
maskGR[maskGR['count'] > 10].sort_values(
    by='mean',
    ascending=False
)
# Найдите сочетание жанров (genres) за 2018 год, которое имеет наибольший средний рейтинг (среднее по столбцу rating), 
# и при этом число выставленных ему оценок (количество значений в столбце rating) больше 10.
```

```python
rat_END_copy['date'] = pd.to_datetime(rat_END_copy['date'])
rat_END_copy['year_rating'] = rat_END_copy['date'].dt.year
pivot = rat_END_copy.pivot_table(
    index='year_rating',
    columns='genres',
    values='rating',
    aggfunc='mean'
)
display(pivot['Animation|Children|Mystery'])
```
